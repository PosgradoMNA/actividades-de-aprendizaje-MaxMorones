##Selección y limpieza de los Datos en Python


```python
input= 'https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv'
```

```python
import pandas as pd
df = pd.read_csv(input)
df
```

Procedemos a revisar la base de datos, con la finalidad de descubrir si hay datos perdidos 

```python
df.isnull().values.any()
```

Con el comando introducido anteriormente podemos confirmar que si tenemos dato perdidos.
Ahora procedemos a introducir el comando para verificar si tenemos datos perdidos en las columnas.

```python
df.isnull().any()
```

Con el comando introducido anteriormente, podemos comprobar que si tenemos valores perdidos.

Ya que no podemos trabajar con datos nulos o datos perdidos, procedemos a implementar diferentes soluciones las cuales pueden ayudarnos a solucionar el problema que se nos está presentando.

# Solucion 1: Eliminar todos los renglones si estos presentan datos perdidos.

---




```python
ndf = df.copy()
ndf.dropna(inplace = True)
ndf
```

Como podemos apreciar, fueron eliminados o removidos 42 renglones los cuales contaban con datos perdidos.

Otra solución que podemos introducir consiste en la eliminacion de la columna, si esta presenta datos perdidos 

Solucion 1.2: Eliminacion de la columna si esta presenta datos perdidos

```python
ndf = df.copy()
ndf.dropna(inplace = True, axis = 1)
ndf
```

Solución 1.3: Eliminar los renglones si todos los valores se encuentran perdidos.

```python
ndf = df.copy()
ndf.dropna(how = 'all', inplace = True)
ndf
```

Para este caso no se elimina ningun renglon, ya que no existe un renglon que cuenten con todos los datos perdidos.

Excepción 1: Si existen al menos 3 datos en un renglon, estos no se eliminaran.

```python
ndf = df.copy()
ndf.dropna(thresh = 3, inplace = True)
ndf
```

Para este caso no se elimino ningun renglon y  ahora sabemos cuales son los renglones que cuentan con más de tres datos.
